```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xaxis= np.arange(1,9)
```

```python
yaxis=np.arange(3,27,3)
```

```python
plt.plot(xaxis, yaxis)
```

```python
plt.plot(xaxis, yaxis, color='green')
plt.show()
```

```python
#plotting with labels
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('Plot graph')
plt.plot(xaxis,yaxis,color='red')
plt.show()
```

```python
X= np.linspace(-4,4,50)
Y= X**2
plt.plot(X,Y)
plt.show()
```

```python
plt.plot(X, np.sin(X), 'r.')
plt.plot(X, np.cos(X), 'g--')
plt.show()
```

```python
#subplots
X=np.arange(1,9)
Y=np.arange(3,27,3)
plt.plot(X,Y)
```

```python
#Creating subplots
plt.subplot(2,2,1)
plt.plot(X,Y, 'red')
plt.subplot(2,2,2)
plt.plot(Y,Y, 'green')
plt.subplot(2,2,3)
plt.plot(X,X, 'blue')
plt.subplot(2,2,4)
plt.plot(Y,X, 'yellow')
```

```python
plt.subplot(2,2,1)
plt.plot(X,X*X,'red')
plt.subplot(2,2,2)
plt.plot(X,np.sqrt(X),'blue')
plt.subplot(2,2,3)
plt.plot(X,np.exp(X),'green')
plt.subplot(2,2,4)
plt.plot(X, np.log10(X),'yellow')
```

```python
#Subplots using Figure, Axes
Figure, Axes= plt.subplots(nrows=2, ncols=1)
Axes[0].plot(X,Y)
Axes[0].set_title('X-Y graph')
Axes[1].plot(X, X**2)
Axes[1].set_title('Square Graph')
```

```python
Figure2, Axes2= plt.subplots(nrows=1, ncols=2, figsize=(14,6))
Axes2[0].plot(Y, Y**3, 'orange')
Axes2[0].set_title('Cube graph')
Axes2[1].plot(X, X**2, 'green')
Axes2[1].set_title('Square graph')
```

```python
Figure2.savefig('Figure1.png')
Figure2.savefig('Figure1.pdf')
```

```python
#multiple graphs on same axes- method1
Figure3, Axes3= plt.subplots(figsize=(15,6))
Axes3= Figure3.add_axes([0,0,1,1])
Axes3.plot(X,X**2, label='X square',color='red')
Axes3.plot(Y, Y**0.5, label='Y square root',color='green')
Axes3.plot(X,Y, label='X-Y', color='orange')
Axes3.plot(Y, np.sin(Y), label='Sin graph',color='yellow')
Axes3.legend()
```

```python
#multiple graphs on same axes- method2- without using subplots
F3= plt.figure(figsize=(15,6))
A3= F3.add_axes([0,0,1,1])
A3.plot(X,X**2, label='X square',color='red')
A3.plot(Y, Y**0.5, label='Y square root',color='green')
A3.plot(X,Y, label='X-Y', color='orange')
A3.plot(Y, np.sin(Y), label='Sin graph',color='yellow')
A3.legend()
```

```python
#Customizing graphs
Figure5= plt.figure()
Axes5= Figure5.add_axes([0,0,1,1])
Axes5.plot(X,Y**2, color='green', lw=5, ls='--',marker='D',markersize=10,
           markerfacecolor='blue', markeredgecolor='black',markeredgewidth=2)
```
