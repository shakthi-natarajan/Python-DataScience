```python
#Find the number of bike shares when the weather is hottest 
#hottest
MaxTemp=DF1['Temperature'].max()
BikeShareHD= DF1[DF1['Temperature']== MaxTemp]
BikeShareHD
BikeShareHD['Count'].sum()
```

```python
#Find the number of bike shares when the weather is coldest
MinTemp=DF1['Temperature'].min()
BikeShareCD=DF1[DF1['Temperature']== MinTemp]
BikeShareCD['Count'].sum()
```

```python
#Find the number of bike shares when the wind speed is highest
HighestWS=DF1['Wind Speed'].max()
BikeShareHW=DF1[DF1['Wind Speed']==HighestWS]
BikeShareHW['Count'].sum()
```

```python
#Find the lowest number of bike shares in winter
DF1_Group_Season= DF1.groupby('Season')
Group_Winter=DF1_Group_Season.get_group(3)
Group_Winter['Count'].min()
```

```python
#another way of finding lowest number of bike share in winter
DF1_Group_Season= DF1.groupby('Season')
DF1_Group_Season.get_group(3)['Count'].min()
```

```python
#Find the highest number of bike shares in autumn
Group_Autumn=DF1_Group_Season.get_group(2)
Group_Autumn['Count'].max()
```

```python
#Find the average humidity in spring
Group_Spring=DF1_Group_Season.get_group(0)
Group_Spring['Humidity'].mean()
```

```python
#summer
Group_Summer=DF1_Group_Season.get_group(1)
```

```python
#Another way of scatter plot
x= DF1['Temperature']
y= DF1['Count']
Figure, Axes= plt.subplots(figsize=(140,60))
Axes.scatter(x,y)
#plt.show()
```

```python
#Create a pie chart that showed the distribution of bike shares by season.
Labels=['Spring', 'Summer', 'Fall', 'Winter']
Values=[Group_Spring['Count'].sum(),
    Group_Summer['Count'].sum(), 
    Group_Autumn['Count'].sum(),
    Group_Winter['Count'].sum()]
Figure, Axes= plt.subplots()
Axes.pie(Values, labels=Labels, autopct='%1.1f%%')
plt.show()
```
