```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
EPL=pd.read_csv('2006-2018 EPL stats.csv')
EPL
```

```python
#rows
len(EPL)
EPL.count()
```

```python
EPL.count()
EPL['season'].value_counts()
```

```python
#find the most successful season of Manchester United and keep it
EPL_group= EPL.groupby(['team'])
Group_Manchester= EPL_group.get_group('Manchester United')
MU_Success=Group_Manchester[Group_Manchester['wins']== Group_Manchester['wins'].max()]
```

```python
MU=MU_Success.groupby('season').get_group('2008-2009')
MU
```

```python
#find the most successful season of Manchester City and keep it
EPL_Team= EPL.groupby('team')
EPL_MC= EPL_Team.get_group('Manchester City')
MC=EPL_MC[EPL_MC['wins']== EPL_MC['wins'].max()]
MC
```

```python
#find the most successful season of Liverpool and keep it
EPL_L= EPL_Team.get_group('Liverpool')
LP=EPL_L[EPL_L['wins'].max()== EPL_L['wins']]
LP
```

```python
#find the most successful season of Chelsea and keep it
EPL_C=EPL_Team.get_group('Chelsea')
Chelsea=EPL_C[EPL_C['wins']==EPL_C['wins'].max()]
Chelsea
```

```python
#find the most successful season of Arsenal and keep it
EPL_A=EPL_Team.get_group('Arsenal')
Arsenal=EPL_A[EPL_A['wins']==EPL_A['wins'].max()]
Arsenal=Arsenal.groupby('season').get_group('2007-2008')
Arsenal
```

```python
#find the most successful season of Tottenham and keep it
EPL_T=EPL_Team.get_group('Tottenham Hotspur')
TH=EPL_T[EPL_T['wins']==EPL_T['wins'].max()]
TH
```

```python
#create a new dataframe and add these teams
DF_merged=pd.concat([MU,MC,LP,Chelsea,Arsenal,TH])
DF_merged
```

```python
#find the most successful team in the new dataframe
Winner=DF_merged[DF_merged['wins'].max()==DF_merged['wins']]
Winner
```

```python
#find the team that lost the most in the new dataframe
DF_merged[DF_merged['losses'].max()==DF_merged['losses']]
```

```python
#find the team with the most accurate shots on target in 2016
DF_Y=EPL.groupby('season')
DF_2016=pd.concat([DF_Y.get_group('2016-2017',DF_Y.get_group('2015-2016')])
DF_2016[DF_2016['ontarget_scoring_att']==DF_2016['ontarget_scoring_att'].max()]
```

```python
#Add a new column named “Penalty success rate” and calculate it= (Penalty saved/penalty conceded)*100
DF_merged['Penalty success rate']=(DF_merged.penalty_save/DF_merged.penalty_conceded)*100
DF_merged
```

```python
#Find the most successful team on penalty saving at new data frame.
DF_merged[DF_merged['Penalty success rate'].max()==DF_merged['Penalty success rate']]
```

```python
#Create a bar chart that showed the distribution of teams by points or wins
x=DF_merged['team']
y=DF_merged['wins']
Figure, Axes=plt.subplots(figsize=(14,8))
Axes.set_xlabel('Top Teams')
Axes.set_ylabel('Wins')
Axes.bar(x,y)
plt.show()
```

```python
#Create a pie chart that showed the distribution of teams by goal attempts- Total scoring att
Labels=DF_merged['team']
values=DF_merged['total_scoring_att']
Figure,Axes=plt.subplots(figsize=(14,10))
Axes.pie(values, labels=Labels, autopct='%1.1f%%', explode=[0,0.1,0.1,0,0.1,0.1])
Axes.legend()
plt.show()
```
